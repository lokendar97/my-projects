```python
def parse_star_count(stars_str):
    if stars_str[-1] == 'k':
        return int(float(stars_str[:-1]) * 1000)
    return int(stars_str)
```

```python
def get_topic_page(topic_url):
    # download the page
    response = requests.get(topic_url)
    
    # check if its sucessfully loaded or not
    if response.status_code != 200:
        raise Exception('Failed to load page{}'.format(topic_url))
    
    #parsing using beautifulsoup
    topic_doc = bs4.BeautifulSoup(response.text,'html.parser')
    return topic_doc

def get_repo_info(h3_tag,star_tag):
    # returns all the required info about a repository
    a_tags = h3_tag.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href']
    stars = parse_star_count(star_tag.text)
    return username,repo_name,repo_url,stars

def get_topic_repo(topic_doc):
    # get h3_tags containing repo info
    h3_selection_class = "f3 color-fg-muted text-normal lh-condensed"
    repo_tags = topic_doc.find_all('h3',{'class': h3_selection_class}) 
    
    # get star tags
    star_tags = topic_doc.find_all('span',{'class':'Counter js-social-count'})
    
    
    topics_repo_dict = {
    'user_name' : [],
    'repo_name' : [],
    'stars'     : [],
    'repo_url'  : []
    }
    

    


    for i in range(len(repo_tags)):
        repo_info = get_repo_info(repo_tags[i],star_tags[i])
        topics_repo_dict['user_name'].append(repo_info[0])
        topics_repo_dict['repo_name'].append(repo_info[1])
        topics_repo_dict['stars'].append(repo_info[3])
        topics_repo_dict['repo_url'].append(repo_info[2])

    return pd.DataFrame(topics_repo_dict)
```

```python
def get_topic_titles(doc):
    selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'
    topic_title_tags = doc.find_all('p', {'class': selection_class})
    topic_titles = []
    for tag in topic_title_tags:
        topic_titles.append(tag.text)
    return topic_titles
```
